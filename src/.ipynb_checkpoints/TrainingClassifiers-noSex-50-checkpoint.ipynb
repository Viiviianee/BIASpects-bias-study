## SVM

```python
import numpy as np
import pandas as pd
import random as random
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from decimal import *
import warnings
warnings.filterwarnings("ignore")

SVMs=[]
name='SVM'
df=pd.read_pickle(filepath + '/Group19')

# Extracting information from Dataframe for all 550 datasets for one experimental group
group=df.loc[:,'Exp. Group'].iloc[0]
ratio=df.loc[:,'Ratio Men'].iloc[0]
prob_men=df.loc[:,'Probability Men'].iloc[0]
prob_women=df.loc[:,'Probability Women'].iloc[0]
diff_prob=df.loc[:,'Difference Probabilities'].iloc[0]

for j in range(1,551):
    # Extract information for every set of the 550 datasets
    df_sub=df[df['# Dataset']==j]
    dataset=df_sub.loc[:,'# Dataset'].iloc[0]
    seed=df_sub.loc[:,'Seed'].iloc[0]
    # Drop unecessary columns for classifier
    df_sub=df_sub.drop(['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities'], axis=1)
    # Split dataset. Use first 1000 samples for training
    df_train=df_sub.iloc[0:1000]
    # Fit scaler
    scaler=StandardScaler()
    X_train=scaler.fit_transform(df_train.drop(['Sex','Label'],axis=1))
    y_train=df_train['Label']
    # Train SVM
    SVM=SVC(kernel='rbf',C=1.5)
    SVM.fit(X_train,y_train)
    #append
    SVMs.append(SVM)

results_df=pd.DataFrame(columns=['Classifier','Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities','Confusion Matrix Men', 'Confusion Matrix Women'])
for i in range(1,149):
    df=pd.read_pickle(filepath + '/Group{}'.format(i))
    # Extracting information from Dataframe for all 550 datasets for one experimental group
    group=df.loc[:,'Exp. Group'].iloc[0]
    ratio=df.loc[:,'Ratio Men'].iloc[0]
    prob_men=df.loc[:,'Probability Men'].iloc[0]
    prob_women=df.loc[:,'Probability Women'].iloc[0]
    diff_prob=df.loc[:,'Difference Probabilities'].iloc[0]
    for j in range(1,551):
        # Get one SVM out of the list
        SVM=SVMs[j-1]
        # Extract information for every set of the 550 datasets
        df_sub=df[df['# Dataset']==j]
        dataset=df_sub.loc[:,'# Dataset'].iloc[0]
        seed=df_sub.loc[:,'Seed'].iloc[0]
        # Drop unecessary columns for classifier
        df_sub=df_sub.drop(['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities'], axis=1)
        # Split dataset. Use second 1000 samples for testing
        df_test=df_sub.iloc[1000:2000]
        # Group test data by sex
        X_test_m=scaler.transform(df_test.drop(['Sex','Label'],axis=1)[df_test['Sex']==0])
        X_test_f=scaler.transform(df_test.drop(['Sex','Label'],axis=1)[df_test['Sex']==1])
        y_test_m=df_test['Label'][df_test['Sex']==0]
        y_test_f=df_test['Label'][df_test['Sex']==1]
        # Predict for both groups and creating confusion matrix of the results
        y_predict_m=SVM.predict(X_test_m)
        y_predict_f=SVM.predict(X_test_f)
        matrix_m=np.zeros((2,2),int)
        matrix_f=np.zeros((2,2),int)
        for true_label,predicted_label in zip(y_test_m,y_predict_m):
            if true_label ==0:
                if predicted_label==0:
                    matrix_m[0,0]+=1
                else:
                    matrix_m[0,1]+=1
            else:
                if predicted_label==0:
                    matrix_m[1,0]+=1
                else:
                    matrix_m[1,1]+=1
                    
        for true_label,predicted_label in zip(y_test_f,y_predict_f):
            if true_label ==0:
                if predicted_label==0:
                    matrix_f[0,0]+=1
                else:
                    matrix_f[0,1]+=1
            else:
                if predicted_label==0:
                    matrix_f[1,0]+=1
                else:
                    matrix_f[1,1]+=1
        # Save confusion matrix in dataframe
        result=pd.DataFrame({'Classifier':name,'Exp. Group':group,'# Dataset':dataset,'Seed':seed,'Ratio Men':ratio,'Probability Men':prob_men,'Probability Women':prob_women,'Difference Probabilities':diff_prob,'Confusion Matrix Men':[matrix_m],'Confusion Matrix Women':[matrix_f]})
        results_df=results_df.append(result,ignore_index=True)
        
    print('Exp. Group {} finished'.format(i))

results_df.to_pickle('./DF_Results_SVM_noSex_50.pkl')
```

## Decision Tree

```python
# Same code as for SVM except the type of classifier

import pandas as pd
import numpy as np
import random as random
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from decimal import *
import warnings
warnings.filterwarnings("ignore")

Trees=[]
name='Tree'
df=pd.read_pickle(filepath + '/Group19')
group=df.loc[:,'Exp. Group'].iloc[0]
ratio=df.loc[:,'Ratio Men'].iloc[0]
prob_men=df.loc[:,'Probability Men'].iloc[0]
prob_women=df.loc[:,'Probability Women'].iloc[0]
diff_prob=df.loc[:,'Difference Probabilities'].iloc[0]
for j in range(1,551):
    df_sub=df[df['# Dataset']==j]
    dataset=df_sub.loc[:,'# Dataset'].iloc[0]
    seed=df_sub.loc[:,'Seed'].iloc[0]
    df_sub=df_sub.drop(['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities'], axis=1)
    df_train=df_sub.iloc[0:1000]
    scaler=StandardScaler()
    X_train=scaler.fit_transform(df_train.drop(['Sex','Label'],axis=1))
    y_train=df_train['Label']
    # Creating Decision Tree
    tree = DecisionTreeClassifier(min_samples_split=0.01, min_samples_leaf=0.01)
    tree.fit(X_train,y_train)

    Trees.append(tree)

results_df=pd.DataFrame(columns=['Classifier','Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities','Confusion Matrix Men', 'Confusion Matrix Women'])
for i in range(1,149):
    df=pd.read_pickle(filepath + '/Group{}'.format(i))
    group=df.loc[:,'Exp. Group'].iloc[0]
    ratio=df.loc[:,'Ratio Men'].iloc[0]
    prob_men=df.loc[:,'Probability Men'].iloc[0]
    prob_women=df.loc[:,'Probability Women'].iloc[0]
    diff_prob=df.loc[:,'Difference Probabilities'].iloc[0]
    for j in range(1,551):
        tree=Trees[j-1]
        df_sub=df[df['# Dataset']==j]
        dataset=df_sub.loc[:,'# Dataset'].iloc[0]
        seed=df_sub.loc[:,'Seed'].iloc[0]
        df_sub=df_sub.drop(['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Difference Probabilities'], axis=1)
        df_test=df_sub.iloc[1000:2000]
        X_test_m=scaler.transform(df_test.drop(['Sex','Label'],axis=1)[df_test['Sex']==0])
        X_test_f=scaler.transform(df_test.drop(['Sex','Label'],axis=1)[df_test['Sex']==1])
        y_test_m=df_test['Label'][df_test['Sex']==0]
        y_test_f=df_test['Label'][df_test['Sex']==1]

        y_predict_m=tree.predict(X_test_m)
        y_predict_f=tree.predict(X_test_f)
        matrix_m=np.zeros((2,2),int)
        matrix_f=np.zeros((2,2),int)
        for true_label,predicted_label in zip(y_test_m,y_predict_m):
            if true_label ==0:
                if predicted_label==0:
                    matrix_m[0,0]+=1
                else:
                    matrix_m[0,1]+=1
            else:
                if predicted_label==0:
                    matrix_m[1,0]+=1
                else:
                    matrix_m[1,1]+=1
                    
        for true_label,predicted_label in zip(y_test_f,y_predict_f):
            if true_label ==0:
                if predicted_label==0:
                    matrix_f[0,0]+=1
                else:
                    matrix_f[0,1]+=1
            else:
                if predicted_label==0:
                    matrix_f[1,0]+=1
                else:
                    matrix_f[1,1]+=1
        
        result=pd.DataFrame({'Classifier':name,'Exp. Group':group,'# Dataset':dataset,'Seed':seed,'Ratio Men':ratio,'Probability Men':prob_men,'Probability Women':prob_women,'Difference Probabilities':diff_prob,'Confusion Matrix Men':[matrix_m],'Confusion Matrix Women':[matrix_f]})
        results_df=results_df.append(result,ignore_index=True)
        
    print('Exp. Group {} finished'.format(i))

results_df.to_pickle('./DF_Results_Tree_noSex_50.pkl')
```
