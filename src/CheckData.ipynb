```python
import pandas as pd
import math
```

```python
eps = 1.0
while eps + 1 > 1:
    eps /= 2
eps *= 2

for i in range(1,149):
    df = pd.read_pickle('/Users/viviane/Documents/Uni/4.Semester/RDS2019/Daten/group{}'.format(i))
    
    print('Exp. Group {}'.format(i))
    
    prob_m = df.at[0, 'Probability Men']
    prob_w = df.at[0, 'Probability Women']
    
    for a in range(1, 551):      
        df_dataset = df[df['# Dataset']==a]
        
        count_w1 = len((df_dataset.query('Sex == 1')).query('Label == 1'))
        count_w0 = len(df_dataset.query('Sex ==1 & Label ==0'))
        count_m1 = len(df_dataset.query('Sex ==0 & Label ==1'))
        count_m0 = len(df_dataset.query('Sex ==0 & Label==0 '))
        
        pw = count_w1/(count_w1+count_w0)
        pm = count_m1/(count_m1+count_m0)
        
    if ((math.fabs(float(prob_m) - pm) > eps)&(math.fabs(float(prob_w) - pw) > eps)):
        print('Anzahl Daten ' + str(count_w1 + count_w0 + count_m1 + count_m0))
        print('Prob. Men = '+ str(prob_m) + ' im Datenset: ' + str((count_m1/(count_m1 + count_m0))))
        print('Prob. Women = '+ str(prob_w) + ' im Datenset:'+ str((count_w1/(count_w1 + count_w0))))
        
```
