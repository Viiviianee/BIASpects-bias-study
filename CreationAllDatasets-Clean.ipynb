```python
from decimal import * 
import random
import pandas as pd
```

```python
def createSample(group,sex):
    
    if sex=='M':
        height=random.gauss(178.4,7.6)
        ability1=random.gauss(MEANS[format(prob_men,'.2f')][0],STD_ABILITY1)
        ability2=random.gauss(MEANS[format(prob_men,'.2f')][1],STD_ABILITY2)
        ability1_noise=ability1+random.gauss(0,STD_NOISE)
        ability2_noise=ability2+random.gauss(0,STD_NOISE)
    else:
        height=random.gauss(164.7,7.1)
        ability1=random.gauss(MEANS[format(prob_men+diff_prob, '.2f')][0],STD_ABILITY1)
        ability2=random.gauss(MEANS[format(prob_men+diff_prob, '.2f')][1],STD_ABILITY2)
        ability1_noise=ability1+random.gauss(0,STD_NOISE)
        ability2_noise=ability2+random.gauss(0,STD_NOISE)

    if ability1+ability2>=40:
        label=1
    else:
        label=0
    
    
    sample=[group,numb_dataset,seed,ratio,prob_men,prob_women,diff_prob,ability1_noise,ability2_noise,height,sex,label]
    return sample


def cleanDataSet(dataList, prob, ratio):
    n1=1000
    droplist= []
    (count_0, count_1) = (0,0)
    
    for row in dataList:
        if (row[11] == 1):
            count_1 += 1
            if (count_1 > round(n1*ratio*prob)):
                droplist.append(row)
            else:
                continue
        else:
            count_0 += 1
            if (count_0 > round((n1*ratio*(1-prob)))):
                droplist.append(row)
            else:
                continue

    for d in droplist:
        dataList.remove(d)

    return dataList



#defining factors
RATIO_MEN=[0.5,0.65,0.8,0.95] 
PROB_MEN=[0.05,0.20,0.35,0.5,0.65,0.8,0.95]
DIFF_PROB=[-0.45,-0.30,-0.15,0,0.15,0.30,0.45]
STD_ABILITY1=4
STD_ABILITY2=3
MEANS={'0.05':[20.88,10.88], '0.20':[22.9,12.9], '0.35':[24.035,14.035], '0.50':[25,15], '0.65':[25.965,15.965], '0.80':[27.1,17.1], '0.95':[29.12,19.12]}
STD_NOISE=2.5
NUMBER_DATASETS=550

n=2000


#creating dataset
counter=0
for ratio in RATIO_MEN:
    ratio=round(Decimal(ratio),2)
    for prob_men in PROB_MEN:
        prob_men=round(Decimal(prob_men),2)
        for diff_prob in DIFF_PROB:
            diff_prob=round(Decimal(diff_prob),2)
            if 0<prob_men+diff_prob<1:
                prob_women=round(Decimal(prob_men+diff_prob),2)
                df_list_M=[]
                df_list_W=[]
                df_list_Mtest=[]
                df_list_Wtest=[]
                df_list=[]
                for numb_dataset in range(1,NUMBER_DATASETS+1):
                    seed=random.randint(0,10000)
                    random.seed(seed)
                    
                    for number in range(0,2):
                        if (number == 0):
                            for i in range(0,int(n*ratio)):
                                df_list_M.append(createSample(counter+1,'M'))
                            for i in range(0,int(n*(1-ratio))):
                                df_list_W.append(createSample(counter+1,'F'))
                        else:
                            for i in range(0,int(n*ratio)):
                                df_list_Mtest.append(createSample(counter+1,'M'))
                            for i in range(0,int(n*(1-ratio))):
                                df_list_Wtest.append(createSample(counter+1,'F'))
                               
                    df_list.extend(cleanDataSet(df_list_M, prob_men, ratio))
                    df_list.extend(cleanDataSet(df_list_W, prob_women, (1-ratio)))
                    df_list.extend(cleanDataSet(df_list_Mtest, prob_men, ratio))
                    df_list.extend(cleanDataSet(df_list_Wtest, prob_women, (1-ratio)))
                    
                print('Exp. Group {}'.format(counter+1))
                datasets=pd.DataFrame(df_list,columns=['Exp. Group', '# Dataset', 'Seed', 'Ratio Men', 'Probability Men', 'Probability Women', 'Difference Probabilities','Ability Score 1', 'Ability Score2', 'Height', 'Sex', 'Label'])
                datasets.to_pickle('./Group{}'.format(counter+1))
                
                counter+=1
    
```
