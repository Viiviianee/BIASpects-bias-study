```python
import pickle

possible_accuracys=[]
```

```python
def accurater(dataname):
    with open(dataname,'rb') as f:
        data=pickle.load(f)
    true=0
    false=0
    for i in range(0,len(data)):
        if data['Ability Score 1'][i] + data['Ability Score2'][i] >=40:
            if data['Label'][i]==1:
                true+=1 
            else:
                false+=1
        else:
            if data['Label'][i]==0:
                true+=1
            else:
                false+=1       
    print(true)
    print(false)
    print(true/(true+false))
    possible_accuracys.append(true/(true+false))
    
    return (true/(true+false))
```

```python
for i in range(1,149):
    accurater('Group{}'.format(i))
    print('Group{} is finished'.format(i))
    print(possible_accuracys[i-1])
```

```python
import numpy as np
print(np.mean(possible_accuracys))
print(np.std(possible_accuracys))
```

```python
import pandas as pd
accs=pd.DataFrame({'possible accuracys':possible_accuracys})
accs.to_pickle('possibleacc.pkl')
```
