```python
import numpy as np
import pandas as pd
import random as random
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings("ignore")

results_df=pd.DataFrame(columns=['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women','Confusion Matrix Men', 'Confusion Matrix Women'])
for i in range(1,149):
    df=pd.read_pickle('./group{}'.format(i))
    df['Sex']=df['Sex'].map({'M':0,'F':1})
    group=df.loc[:,'Exp. Group'].iloc[0]
    ratio=df.loc[:,'Ratio Men'].iloc[0]
    prob_men=df.loc[:,'Probability Men'].iloc[0]
    prob_women=df.loc[:,'Probability Women'].iloc[0]
    for j in range(1,551):
        df_sub=df[df['# Dataset']==j]
        dataset=df_sub.loc[:,'# Dataset'].iloc[0]
        seed=df_sub.loc[:,'Seed'].iloc[0]
        df_sub=df_sub.drop(['Exp. Group', '# Dataset', 'Seed','Ratio Men','Probability Men', 'Probability Women'], axis=1)
        df_train=df_sub.iloc[0:999]
        df_test=df_sub.iloc[1000:1999]
        scaler=StandardScaler()
        X_train=scaler.fit_transform(df_train.drop('Label',axis=1))
        X_test_m=scaler.transform(df_test.drop('Label',axis=1)[df_test['Sex']==0])
        X_test_f=scaler.transform(df_test.drop('Label',axis=1)[df_test['Sex']==1])
        y_train=df_train['Label']
        y_test_m=df_test['Label'][df_test['Sex']==0]
        y_test_f=df_test['Label'][df_test['Sex']==1]

        SVM=SVC(kernel='linear',C=1)
        SVM.fit(X_train,y_train)
        y_predict_m=SVM.predict(X_test_m)
        y_predict_f=SVM.predict(X_test_f)
        matrix_m=confusion_matrix(y_test_m,y_predict_m)
        matrix_f=confusion_matrix(y_test_f,y_predict_f)
        
        result=pd.DataFrame({'Exp. Group':group,'# Dataset':dataset,'Seed':seed,'Ratio Men':ratio,'Probability Men':prob_men,'Probability Women':prob_women,'Confusion Matrix Men':[matrix_m],'Confusion Matrix Women':[matrix_f]})
        results_df=results_df.append(result,ignore_index=True)
    print('Exp. Group {} finished'.format(i))

results_df.to_pickle('./DF_Results')
```
